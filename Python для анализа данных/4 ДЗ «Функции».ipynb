### Домашнее задание к лекции "Функции"

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}      

def start():
    command_map = {
        'p': 'print_owner',
        's': 'print_directories',
        'l': 'print_info',
        'ads': 'add_directory',
        'ds': 'del_directory',
        'ad': 'add_document',
        'd': 'del_document',
        'm': 'move_document'              
    }
    
    command = ''
    
    while command != 'q':
        print('Введите команду:')
        command = input()
        
        if command in command_map:
            globals()[command_map[command]]()    
            
def print_owner():
    doc_num = input_doc_number()    
    document = get_document(doc_num)
    
    if document:
        print(f'Владелец документа: { document["name"] }')        
    else:
        print('Документ не найден в базе.')
              
def print_directories():
    doc_num = input_doc_number()
    directory = get_directory(doc_num)
    
    if directory:
        print(f'Документ хранится на полке: {directory}')
    else:
        print('Документ не найден в базе')
        
def print_info():        
    for doc in documents:
        doc_type, doc_num, doc_name = doc.values()
        print(f'№: {doc_num}, тип: {doc_type}, владелец: {doc_name}, полка хранения: {get_directory(doc_num)}')
        
def add_directory():
    dir_num = input_doc_directory()
    
    if dir_num in directories:
        print(f'Такая полка уже существует. Текущий перечень полок: {get_directories_list()}.')
    else:
        directories[dir_num] = []
        print(f'Полка добавлена. Текущий перечень полок: {get_directories_list()}.')
        
def del_directory():
    dir_num = input_doc_directory()
    
    if dir_num not in directories:
        print(f'Такой полки не существует. Текущий перечень полок: {get_directories_list()}.')
        return
    
    if directory_is_empty(dir_num):
        del directories[dir_num]
        print(f'Полка удалена. Текущий перечень полок: {get_directories_list()}.')
    else:
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {get_directories_list()}.') 
              
def add_document():
    doc_num = input_doc_number()
    
    print('Введите тип документа:')
    doc_type = input()
    
    print('Введите владельца документа:')
    doc_owner = input()
    
    print('Введите полку для хранения:')
    doc_dir = input()
    
    if doc_dir in directories:
        add_doc_in_documents({ 'type': doc_type, 'number': doc_num, 'name': doc_owner })
        add_doc_in_directory(doc_dir, doc_num)       
        print('Документ добавлен.')        
    else:
        print('Такой полки не существует. Добавьте полку командой ads.')
    
    print('Текущий список документов:')
    print_info()
    
def del_document():
    doc_num = input_doc_number()
    
    doc_index = get_document_index(doc_num)
    doc_dir = get_directory(doc_num)
    
    if doc_index and doc_dir:
        del_doc_in_documents(doc_index)
        del_doc_in_directory(doc_dir, doc_num)
        print('Документ удален.')
    else:
        print('Документ не найден в базе.')
    
    print('Текущий список документов:')
    print_info()
    
def move_document():
    doc_num = input_doc_number()
    document = get_document(doc_num)
              
    if document == None:
        print('Документ не найден в базе.')
        print('Текущий список документов:')
        print_info()
        return
              
    dir_num = input_doc_directory()
    
    if dir_num not in directories:
        print(f'Такой полки не существует. Текущий перечень полок: {get_directories_list()}.')
        return        

    doc_index = get_document_index(doc_num)

    del_doc_in_documents(doc_index)
    del_doc_in_directory(get_directory(doc_num), doc_num)

    add_doc_in_documents(document)
    add_doc_in_directory(dir_num, doc_num)  

    print('Документ перемещён.')        
    print('Текущий список документов:')
    print_info()
              
def input_doc_number():
    print('Введите номер документа:')
    return input()

def input_doc_directory():
    print('Введите номер полки:')
    return input()

def get_document(doc_num): 
    for doc in documents:
        if doc['number'] == doc_num:
            return doc       
        
def get_document_index(doc_num): 
    for i, doc in enumerate(documents):
        if doc['number'] == doc_num:
            return i
        
def get_directory(doc_num): 
    for directory, docs  in directories.items():
        if doc_num in docs:
            return directory
        
def get_directories_list(): 
    return ', '.join(directories.keys())

def directory_is_empty(dir_num): 
    return not directories[dir_num]

def add_doc_in_documents(doc):
    documents.append(doc)
    
def add_doc_in_directory(doc_dir, doc_num):
    if doc_num not in directories[doc_dir]:
        directories[doc_dir] = [*directories[doc_dir], doc_num]
        
def del_doc_in_documents(index):
    del documents[index]
    
def del_doc_in_directory(doc_dir, doc_num):
    new_dir = [*directories[doc_dir]]
    del new_dir[new_dir.index(doc_num)]
    directories[doc_dir] = new_dir 
              
start()
```
