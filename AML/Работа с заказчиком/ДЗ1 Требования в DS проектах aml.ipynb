Возьмите за основу продукт компании “Нетология” - профессию Data Scientist.
Как бы Вы организовали аналитику по данному продукту?

- Сформулируйте 2-3 ключевые метрики продукта
- Какие данные Вам могут помочь в поиске инсайтов?
- Опишите команду для работы с данными для данного продукта
- Какие роли Вам будут нужны?
- Чем они будут заниматься?

<b style="color:blue">Цели</b>

Для начала хорошо бы определиться, какова цель создания DS команды:

Самое очевидное, это увеличение прибыли от продаж профессии. Здесь могут быть разные подзадачи:
- увеличить продажи доп. курсов с помощью рекомендательной системы
- оценить и улучшить качество материалов
- исследовать ценовую политику аналогичных образовательных проектов и построить модель ценообразования
- оценить и улучшить качество работы преподавателей

Чуть менее очевидное, это подготовка квалифицированных DS специалистов (предположим, что мы хотим закрыть деффицит кадров в других своих проектах или просто болеем за идею :)

Исходя из целей <b style="color:blue">Метрики</b> могут быть такими:

- кол-во пользователей, купивших профессию.
Эта метрика скорее оценивает этап внешних продаж (эффективность рекламы, насколько заманчиво наполнение курса, цена и т.п.)

- кол-во студентов, успешно закончивших профессию.
Здесь уже большую роль играет качество материалов, квалификация преподавателей, структура курса и организация обучения

- кол-во студентов, устроившихся на работу после прохождения профессии.
Это основная метрика для программы трудоустройства, но также влияет на оценку качества курсов

<b style="color:blue">Данные</b>, которые можно использовать:


- информация о действиях пользователей на сайте (поисковые запросы, посещение страницы профессии, заказ звонка, покупка)
- отзывы от пользователей, полученные на сайте, в Slack, менеджерами/кураторами через телефонный разговор/почту, 
- отзывы от пользователей, собранные с других ресурсов
- результаты маркетинговых компаний
- информация о скорости и качестве выполнения домашних заданий (время, между принятием ДЗ и выполнением, кол-во возвратов на доработку, продление дед-лайнов)
- информация о семинарах (сколько студентов смотрят онлайн, переносы/отмены)
- обратная связь от преподавателей (опросы, оценки)
- обратная связь от студентов по домашним заданиям и семинарам (отпросы, лайки) 

<b style="color:blue">Структура</b>

Так как профессия является частью большого проекта, я предполагаю, что многие паттерны по сбору/хранению/обработке данных, а также по задачам, которые стоят перед DS специалистами, будут повторяться, поэтому дублировать, дата-инженеров, разработчиков и DS специалистов под отдельную профессию нет смысла. Исходя из этого, я представляю, например, такую структуру:

1. Отдел, который занимается сбором и хранением данных (один на весь проект, а не на профессию). Эти же люди занимаются внедрением готовых моделей, совместно с отделом разработки. 

    Состав: Минимально один дата-инженер. Для крупного проекта, это точно не один человек, несколько дата-инженеров + может быть, архитектор.
    
    
2. Отдел разработки. Он не относится на прямую к DS, но именно туда будут приходить с запросами на добавление ещё одной формы опроса, кнопки "лайк", реализацию отслеживания действий пользователя и т.п.


3. Отдел Data Sience (также один на весь проект, а не на профессию). Можно разделить на две команды с централизованным управлением:

   Одна занимается решением "внешних задач": рекомендательные системы, работа с данными по действиям пользователей, рекламным компаниям, ценообразованию и т.д. 
   
   Вторая занимается решением "внутренних задач": данные по качеству курсов, улучшение кадров и т.п. 

   Состав (минимальный):
   
    - BI-аналитик - один на отдел - занимается общением с заказчиком, крупной аналитикой. Выявляет и формулирует проблемы, отслеживает их решение и понимает, какие результаты заказчик ожидает увидеть от внедрения моделей

    - Аналитик - один на команду - запрашивает данные у дата-инженеров, обрабатывает, исследует, делает отчёты, генерирует идеи

    - Data Scientist - один на команду - обрабатывает и исследует данные, строит модели, проверяет гипотезы
    
   В целом, мне кажется, если задач немного, то Data Scientist может выполнять функции аналитика, но на крупном проекте лучше, если это будут отдельные роли. Также если задач много, то и аналитиков и дата-саентистов в каждой команде будет больше


3. Менеджер продукта: - вот тут вполне может быть один на каждую профессию или направление. Я пока не очень понимаю специфику DS проектов, но из опыта разработки предположу, что хороший продакт - половина успеха. Но только в том случае, если это специалист, который не только двигает таски и по утрам на планёрке спрашивает "ну что, сделали? а почему?", но действительно разбирается в предметной области, понимает, какие проблемы решать технически, а какие менеджментом. Понимает, с какими именно людьми работает, и как с их помощью дойти до результата.
